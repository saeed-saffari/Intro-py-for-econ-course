## Task 1

### Task 1.1: importing covid19 dataset 


```python
corona_dataset = pd.read_csv('time_series_covid19_confirmed_global.csv')
corona_dataset.head()
```

#### Let's check the shape of the dataframe

```python
corona_dataset.shape
```

### Task 1.2: Delete the useless columns

```python
corona_dataset.drop(['Lat', 'Long'], axis = 1, inplace=True)
corona_dataset.head(15)
```

### Task 1.3: Aggregating the rows by the country

```python
corona_dataset_agg = corona_dataset.groupby('Country/Region').sum()
corona_dataset_agg.head(15)
```

```python
corona_dataset_agg.shape
```

### Task 1.4: Visualizing data related to countries (for example China, Italy, Spain and Iran)

```python
plt.figure(figsize = (9,5))
corona_dataset_agg.loc['Iran'].plot(lw = 2)
corona_dataset_agg.loc['Spain'].plot(lw = 2)
corona_dataset_agg.loc['Italy'].plot(lw = 2)
corona_dataset_agg.loc['Iraq'].plot(lw = 2)

plt.legend()
plt.grid()
plt.show()
```

### Task 2.3: find maximum infection rate for all of the countries. 

```python
corona_dataset_agg.head()
```

```python
corona_dataset_agg.index
```

```python
country = list(corona_dataset_agg.index)
#country
```

```python
max_infection = []

for i in country:
    max_infection.append(corona_dataset_agg.loc[i].diff().max())
    
corona_dataset_agg['max_infection_rate'] = max_infection
```

```python
corona_dataset_agg.head()
```

### Task 2.4: create a new dataframe with only needed column 

```python
corona_2 = pd.DataFrame(corona_dataset_agg['max_infection_rate'])
corona_2.head()
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
x = data['Logged GDP per capita']
y = data['max_infection_rate']

sns.scatterplot(x, np.log(y))
```
