## Extra Work

```python
price_AAPL = data.DataReader('AAPL', 'yahoo', start, end)
```

```python
price_AAPL.head()
```

```python
price_AAPL['daily_diffrence'] = price_AAPL['Adj Close'].diff()
price_AAPL.head()
```

```python
price_AAPL['signal'] = np.where(price_AAPL['daily_diffrence'] > 0, 1, 0)
price_AAPL.head()
```

```python
price_AAPL['possition'] = price_AAPL['signal'].diff()
price_AAPL.head()
```

```python
price_AAPL[price_AAPL['possition'] == 1.0].index
```

```python
price_AAPL.loc[price_AAPL['possition'] == 1.0,'Adj Close']
```

```python
plt.figure(figsize = (15,8))

plt.plot(price_AAPL['Adj Close'], c = 'r', lw = 2)

plt.plot(price_AAPL[price_AAPL['possition'] == 1.0].index,
        price_AAPL.loc[price_AAPL['possition'] == 1.0,'Adj Close'], 'o', c = 'g', markersize = 7
        )

plt.plot(price_AAPL[price_AAPL['possition'] == -1.0].index,
        price_AAPL.loc[price_AAPL['possition'] == -1.0,'Adj Close'], 'x', c = 'k', markersize = 7
        )

plt.grid()
plt.plot()
```

### Statemodel 

```python
data = wb.download(country='all', indicator=['NY.GDP.MKTP.CD', 'IT.CEL.SETS.P2'], start = 2017, end = 2017)
```

```python
data = data.dropna()
data.head()
```

```python
data.columns = ['gdp', 'cellphone']
```

```python
data.head()
```

```python
import statsmodels.formula.api as smf
```

```python
mod = smf.ols('cellphone ~ np.log(gdp)', data).fit()
```

```python
print(mod.summary())
```
